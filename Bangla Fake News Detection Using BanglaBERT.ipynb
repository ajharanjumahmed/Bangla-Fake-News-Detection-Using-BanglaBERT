Combining both the dataset with important features

```python
df_real = df1[['headline', 'content', 'label']].copy()
df_fake = df2[['headline', 'content', 'label']].copy()

df_real['text'] = df_real['headline'].astype(str) + " - " + df_real['content'].astype(str)
df_fake['text'] = df_fake['headline'].astype(str) + " - " + df_fake['content'].astype(str)

df_real = df_real[['text', 'label']]
df_fake = df_fake[['text', 'label']]

df_all = pd.concat([df_real, df_fake], ignore_index=True)
```

Reshuffle the combined dataset

```python
df_all = df_all.sample(frac=1, random_state=42).reset_index(drop=True)
```

Import the tokenizer and Bangla BERT Model

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "csebuetnlp/banglabert"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

Split Data into Train and Test Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_all["text"].tolist(),
    df_all["label"].tolist(),
    test_size=0.2,
    random_state=42
)
```

```python
print(len(X_train))
print(len(X_test))
```

Tokenization of train and test texts, keeping maximum length of each news as 512 tokens for fast and accurate training

```python
train_encodings = tokenizer(
    X_train,
    truncation=True,
    padding=True,
    max_length=512
)

test_encodings = tokenizer(
    X_test,
    truncation=True,
    padding=True,
    max_length=512
)
```

Preparing the dataset for Bangla Bert training by converting encodings into PyTorch dataset

```python
import torch

class FakeNewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(int(self.labels[idx]), dtype=torch.long)
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = FakeNewsDataset(train_encodings, y_train)
test_dataset = FakeNewsDataset(test_encodings, y_test)
```

Training the Model

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='./results',
    do_eval=True,
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=50
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()
```

Evaluating the model

```python
results = trainer.evaluate()
results
```

```python
from sklearn.metrics import classification_report
import numpy as np

predictions = trainer.predict(test_dataset)
preds = np.argmax(predictions.predictions, axis=1)

print(classification_report(y_test, preds))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, preds)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
disp.figure_.suptitle("Confusion Matrix")
```

Load the model for prediction or deployment

```python
from transformers import ElectraTokenizer, ElectraForSequenceClassification

loaded_tokenizer = ElectraTokenizer.from_pretrained("./bangla_fake_news_model")
loaded_model = ElectraForSequenceClassification.from_pretrained("./bangla_fake_news_model")

loaded_model.eval()
```

Making the fake news detector function

```python
import torch

def predict_fake_news(text):
    encoding = loaded_tokenizer(
        text,
        truncation=True,
        padding="max_length",
        max_length=512,
        return_tensors="pt"
    )

    outputs = loaded_model(**encoding)
    logits = outputs.logits
    prediction = torch.argmax(logits, dim=1).item()

    label_map = {1: "Real", 0: "Fake"}

    return label_map[prediction]
```
